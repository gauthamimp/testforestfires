## Data Cleaning

```python
## missing values
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

## The daatset is converted into two sets based on region from 122th index,we can make a new column based on the region
## 1:"Bejaia Region Dataset"
## 2:"Sidi-Bel Abbes Region Dataset"
## Add new column with region

```python
df.loc[:122,"Region"]=0 # assigning bejaia region to 0 
df.loc[122:,"Region"]=1 # assigning sidi-bel abbes region to 1
```

```python
df.info() # so now i'll hv additional new feature called as 'Region'
```

```python
df.head() # 'Region' is in the floating value
```

```python
df['Region']=df['Region'].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum() # so now we can see all the null values are removed.
```

```python
df.iloc[[122]]
```

```python
## remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns # so we can see there are extra spaces in 'RH','Ws','Rain' and 'Classes' columns.
```

```python
# fix spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

## change the required columns as integer data type.

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info() # now we can see all the above mentioned columns are converted into int datatype.
```

## changing the other columns to float data datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## let's save the cleaned daatset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

## Exploratory Data Analyis(EDA)

```python
## drop day,month and year
df_copy=df.drop(['day','month','year'],axis=1) # we r dropping them bcz they are not playing any imp roles in predicting the temperature.
```

```python
df_copy.head()
```

```python
## categories in classes
df_copy['Classes'].value_counts()
```

```python
# plot density plot for all features
sns.set_style('whitegrid')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```
